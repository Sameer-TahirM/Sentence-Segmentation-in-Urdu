# 1

### To detect the end of a sentence, they usually end with a verb, or a New line. But if a new line starts and a conjunction come right after, then the sentence is continuing. 

# 2

### I tried to maintain a list of new line words, conjunction words, verb words in my code.

### I try to find a new-line word by first converting the passsage into tokens after normalizing it with urduhack

### Furthermore, once a new-line word is matched, i check that the next word in the tokens is not a conjunction word, apart from this i also check that the next word is not a newline word,I loosely check that the previous word is a verb word.

### If above conditions are met, i combine the current and previous words to make them a sentence and store it in a container.

# 3

#### I count the number of dashes in the original passage, and then i calculate the number of dashes from the algorithm. This would give me P1

#### The compare the end words where dash appears, with the orginal location of the dash, in the corpus, and calculate a Percentage 2

#### Finally I will calculate the actual segment, with my segments, by matching their lengths. (Done individual for each segment). Finally calculate the average of all the segments.

```python
# Function to check if a word is a verb from a list of verb words
def is_verb(word, verb_words):
    for verb in verb_words:
        if verb in word:
            return True
    return False


# Function to check if the word indicates a newline
def is_ending_word(word, index, tokens, newline_words, conjunction_words, verb_words):
    for newline_word in newline_words:
        if (word.endswith(newline_word)) and (index + 1 < len(tokens)) and (tokens[index+1] not in conjunction_words): 
            # Checks if the previous word is a verb word
            if is_verb(tokens[index-1], verb_words):
                return True
            # Checks if the next word is not a newline word
            elif not is_ending_word(tokens[index+1], index + 1, tokens, newline_words, conjunction_words, verb_words):
                # Checks if a comma does not exist in the next word
                if not is_ending_word("،", index + 1, tokens, newline_words, conjunction_words, verb_words):
                    return True
    return False
```

```python
def segmentSentence(passage):

    #List of Conjunction words
    conjunctionWords = ['جنہیں', 'جس', 'جن', 'جو', 'اور', 'اگر', 'اگرچہ', 'لیکن', 'مگر', 'پر', 'یا', 'تاہم', 'کہ', 'کر',
                        'تو', 'گے', 'گی','پہلے','کیونکہ','چونکہ','ورنہ','بلکہ','تاکہ', 'جیسے','چنانچہ']

    #List of Words that denote end of a sentence
    newlineWords = ['کیجیے', 'کیجئے', 'گئیں', 'تھیں', 'ہوں', 'خریدا', 'گے', 'ہونگے', 'گا', 'چاہیے', 'ہوئیں', 'گی',
                        'تھا', 'تھی', 'تھے', 'ہیں', 'ہے','سکے']
    
    # Words that mention Verb words
    verbWords = [ 'کرنا', 'دوڑانہ', 'کھیلنا', 'گھورنا', 'دینا', 'چڑھانا', 'دیکھنا', 'چمکنا', 'جلنا', 'چلنا', 'گھمانا', 'رگڑنا', 'سننا' ]

    #normalizing the passage
    normalizedPassage = normalize(passage)

    print(normalizedPassage)

    #tokenize the normalized passage
    tokens = normalizedPassage.split(' ')

    #empty list to hold sentences
    sentences = []

    #segmenting the sentences
    sentenceFound = False
    sentence = ""

    for index, word in enumerate(tokens):

        #Check for end of sentence words
        if is_ending_word(word, index, tokens, newlineWords, conjunctionWords, verbWords):
            sentenceFound = True

        #Add the current word in sentence container
        if sentence == "":
            sentence += word
        
        else:
            sentence = sentence + " " + word
            if sentenceFound:
                sentence += "۔"

        #Reset the sentence container and add it into sentences container
        if (sentenceFound is True) or ('۔' in word) or ('؟' in word):
            sentenceFound = False
            sentences.append(sentence)
            sentence = ""

    return sentences
```
